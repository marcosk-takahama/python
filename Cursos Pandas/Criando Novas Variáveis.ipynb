## Criando Novas Variáveis

```python
import pandas as pd
dados = pd.read_csv('dados/aluguel_residencial_sem_nulo.csv',sep=";")
```

```python
dados.head(10)
```

```python
#Calculando valor bruto
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
#Calculando valor por metro quadrado
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

```python
#arredondar o valor m2 para 2 casas decimais
dados['Valor m2'] = dados['Valor m2'].round(2)
dados.head(10)
```

```python
#Calculando Valor Bruto por m2, arredondando
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head(10)
```

```python
# Criando uma nova variável para atribuir se é casa ou apartamento
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head(10)
```

## Excluindo Variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_aux.head(10)
```

```python
del dados_aux['Valor Bruto']
```

```python
dados_aux.head()
```

```python
dados_aux.pop('Valor Bruto m2')
dados_aux.head(10)
```

```python
#vamos excluir mais de 1 coluna
#axis =0 -> linha
#axis =1 -> coluna
dados.drop(['Valor Bruto','Valor Bruto m2'],axis=1,inplace=True)
```

```python
dados.head()
```

```python
#salvando em arquivo
dados.to_csv('dados/aluguel_residencial.csv',sep=';',index=False)
```
