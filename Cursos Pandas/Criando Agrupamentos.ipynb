## Criando Agrupamentos

```python
import pandas as pd
dados = pd.read_csv('dados/aluguel_residencial.csv',sep=';')
dados.head(10)
```

```python
#calculando a média de valores
dados['Valor'].mean()
```

```python
dados['Bairro'].unique()
```

```python
#selecionando alguns bairros
bairros=['Copacabana', 'Jardim Botânico', 'Centro', 'Higienópolis',
       'Cachambi', 'Barra da Tijuca', 'Botafogo', 'Grajaú']
bairros
```

```python
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
dados.head(20)
```

```python
#removeu os duplicados
dados['Bairro'].drop_duplicates()
```

```python
#agrupando por bairro
grupo_bairro = dados.groupby('Bairro')
grupo_bairro
```

```python
# criou um objeto tipo groupby
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
#percorrendo os agrupamentos
for bairro, dados in grupo_bairro:
    print(bairro)
```

```python
#percorrendo os DataFrame
# o groupby guarda um dataframe para cada bairro
for bairro, dados in grupo_bairro:
    print(type(dados))
```

```python
#calculando a média de valor por bairro
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro,data.Valor.mean()))
```

```python
#Calculando a média por bairro de uma forma mais simples
grupo_bairro.Valor.mean().round(2)
```

```python
#Calculando a média por bairro e condominio de uma forma mais simples
grupo_bairro[['Valor','Condominio']].mean().round(2)
```

## Estatísticas Descritivas

```python
#função describe retorna alguns valores estatístico
grupo_bairro['Valor'].describe().round(2)
```

```python
#Podemos também trazer os valores separadamente usando o aggregate e renomeando as colunas com o rename
grupo_bairro['Valor'].agg(['min','max','sum']).rename(columns ={'min':'Mínimo','max':'Máximo','sum':'Soma'})
```

```python
#Configurando o matplot
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(20,10))
```

```python
#Plotando o gráfico (média)
fig = grupo_bairro['Valor'].mean().plot.bar(color='blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize':22})
```

```python
#Plotando o gráfico (máximo)
#Podemos verificar que há uma discrepância no valor máximo de Botafogo
fig = grupo_bairro['Valor'].max().plot.bar(color='blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Máximo do Aluguel por Bairro', {'fontsize':22})
```
