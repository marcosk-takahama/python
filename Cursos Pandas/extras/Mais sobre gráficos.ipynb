```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure',figsize=(17,10))
```

```python
dados = pd.read_csv('dados/aluguel.csv',sep=';')
dados.head(10)
```

```python
area =plt.figure()
```

```python
#vamos criar um grafico com 4 graficos (2 linhas e 2 colunas)
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor x Area')

g2.hist(dados.Valor)
g2.set_title('Histograma Valor')

#criando uma amostra de tamanho 100
dados_g3 = dados.Valor.sample(100)
#refazer o indice
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

grupo = dados.groupby('Tipo')['Valor']
#grafico de barras, vai precisar dos rotulos e valores
labels = grupo.mean().index
valores = grupo.mean().values
g4.bar(labels,valores)
g4.set_title('Valor Médio por Tipo')
```
